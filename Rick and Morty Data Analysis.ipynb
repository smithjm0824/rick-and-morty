```python
import requests
base_url="https://rickandmortyapi.com/api/"
character_url=base_url+"character/"
location_url=base_url+"location/"
episode_url=base_url+"episode/"
```

```python
class Character():
    
    def getCharacters(page=1):
        payload = {"page": page}
        return requests.get(character_url, params=payload).json()
    
    def getCharacterById(id=None):
        if id==None:
            raise Exception("Error: We can't find any characters without an Id. Are you Dumb Rick?")
        return requests.get(character_url+str(id)).json()
```

```python
class Location():
    
    def getLocations():
        return requests.get(location_url).json()
    
    def getLocationById(id=None):
        if id==None:
            raise Exception("Error: Don't forget to pass an id of the location.")
        return requests.get(location_url+str(id)).json()
```

```python
Location.getLocationById(1)
```

```python
class Episode():
    
    def getEpisodes(page=1):
        payload = {"page" : page}
        return requests.get(episode_url, params=payload).json()
    
    def getEpisodesById(id=None):
        if id==None:
            raise Exception("Error: This isn't Intergalatic Cable. Enter an ID for the Episode you want.")
        return requests.get(episode_url+str(id)).json()
```

```python
import pandas as pd

titles = pd.read_csv('IMDB TITLE.tsv', delimiter='\t', encoding="utf8")
rick_and_morty_series_id = titles.loc[(titles['originalTitle'] == "Rick and Morty") & (titles['titleType'] == "tvSeries")]['tconst'].item()
if rick_and_morty_series_id != "tt2861424":
    raise Exception('Aww geez Rick... I .. I .. I think you got the wrong series ID: you got {}, it should be tt2861424'.format(rick_and_morty_series_id))
else:
    print("C'mon Morty, real quick adventure. We'll be in and out, five minutes tops")
```

```python
episodes=pd.read_csv('IMDB EPISODE.tsv', delimiter='\t', encoding="utf8")
episode_ids = episodes.loc[episodes['parentTconst'] == rick_and_morty_series_id][['tconst','seasonNumber','episodeNumber']]
```

```python
episode_info = episode_ids.reset_index(drop=True)
display(episode_info)
```

```python
episode_ids = episode_info['tconst'].values
```

```python
ratings = pd.read_csv("IMDB RATINGS.tsv", delimiter="\t", encoding="utf8")
episode_ratings = ratings.loc[ratings['tconst'].isin(episode_ids)]
episode_ratings = episode_ratings.reset_index(drop=True)
display(episode_ratings)
```

```python
episode_table = pd.merge(episode_info, episode_ratings, on="tconst")
```

```python
episode_titles = titles.loc[titles['tconst'].isin(episode_ids)][['tconst','primaryTitle']]
episode_titles = episode_titles.reset_index(drop=True)
episode_table = pd.merge(episode_table, episode_titles, on="tconst")
```

```python
display(episode_table)
```

```python
avgSeriesRating = sum(episode_table['averageRating'])/len(episode_table)
avgSeason1Rating = sum(episode_table.loc[episode_table['seasonNumber']=='1']['averageRating'])/len(episode_table.loc[episode_table['seasonNumber']=='1'])
avgSeason2Rating = sum(episode_table.loc[episode_table['seasonNumber']=='2']['averageRating'])/len(episode_table.loc[episode_table['seasonNumber']=='2'])
avgSeason3Rating = sum(episode_table.loc[episode_table['seasonNumber']=='3']['averageRating'])/len(episode_table.loc[episode_table['seasonNumber']=='3'])

print(round(avgSeriesRating,2))
print(round(avgSeason1Rating,2))
print(round(avgSeason2Rating,2))
print(round(avgSeason3Rating,2))
```

```python
episode_table["id"] = episode_table.apply(lambda row: (row.name + 1), axis=1)
```

```python
characters = []
for i in range(1,26):
    character_details = Character.getCharacters(str(i))
    for character in character_details["results"]:
        characters.append(character)
```

```python
import re
drop_txt = "https://rickandmortyapi.com/api/episode/"

characters_by_id = {}

for character in characters:
    starring_in_episodes = []
    for episode in character["episode"]:
        starring_in_episodes.append(re.split(drop_txt, episode)[1])
    characters_by_id[character['id']] = {'name': character['name'], 
                                         'episodes': starring_in_episodes, 
                                         'species': character['species'],
                                         'gender': character['gender'],
                                         'image': character['image'],
                                         'origin': character['origin'],
                                         'location': character['location']
                                        }
```

```python
episode_list = []
for i in range(1,3):
    episode_details = Episode.getEpisodes(str(i))
    for episode in episode_details["results"]:
        episode_list.append(episode)
```

```python
drop_txt = "https://rickandmortyapi.com/api/character/"
episodes_by_id = {}

for episode in episode_list:
    character_list = []
    for character in episode['characters']:
        character_list.append(re.split(drop_txt, character)[1])
    episodes_by_id[episode['id']] = {'name': episode['name'], \
                                     'characters' : character_list, \
                                     "rating" : episode_table.loc[episode_table['id'] == episode['id']]['averageRating'].item()}
```
